# Calculating Lyapunov exponents in non-elastic polygonal billiards

In this notebook, we will show how Lyapunov exponents can be calculated for inelastic billiards with flat boundaries. We will illustrate this with an equilateral triangle, but the method itself is general and can be applied to any billiard table with flat boundaries.

The phase space for this system is 4-dimensional. Vectors in phase space will be denoted $\Gamma$, and tangent vectors in tangent space as
 $\delta\Gamma$.

## The non-elastic collision rule

To make the collisions non-elastic, we introduce a collision rule that reduces the outgoing angle, $\varphi$,  after the collisions by a factor $\Lambda$:

 $$\varphi = \lambda \theta,$$
 
 where $\theta$ is the incoming angle and  $0 \le \lambda \le 1$.

To illustrate the collision rule, we plot a few collisions of a random initial condition:

```python
# Set up the triangle:
p = Poligono([0,0], 3, 1, regular =  True )
orb = Orbita(p)

lamb = 0.3
num_collisions = 100

orb.flujo(num_collisions, 'd', lamb, calcular_lyapunov=False)

p.draw()
orb.draw_trayectoria()
```

When the collisions are elastic the momentum of the particle is reflected at the boundary:


$$\mathbf{p}' = \mathbf{p} - 2(\mathbf{p}\cdot\mathbf{\hat{n}}) \, \mathbf{\hat{n}} 
= \left[ \mathcal{I} - 2 \mathbf{n} \otimes \mathbf{n} \right] \mathbf{p} =: \mathcal{F}_\text{el}(\mathbf{p}),$$

where $\mathbf{\hat{n}}$ is the unit inwards normal to the boundary at the point of collision.


The transformation due to the inelastic collision can be thought of as an elastic reflection composed with a rotation:

$\mathbf{p}' = \mathcal{F}(\mathbf{p}) = \left[ \mathcal{R}(\mathbf{p}) \circ \mathcal{F}_\mathrm{el} \right] (\mathbf{p}).$

However, since the rotation itself depends on the momentum, this transformation is **not** linear. It has to be expressed as a series of transformations:

$\mathbf{p} \overset{f_1}{\mapsto} \mathbf{p}_{\mathrm{el}} \overset{f_2}{\mapsto} (\mathbf{p}_{\mathrm{el}},\mathbf{p}_{\mathrm{el}}) \overset{f_3}{\mapsto} (\theta_{\mathrm{el}},\mathbf{p}_\mathrm{el} ) \overset{f_4}{\mapsto} (\theta_{\mathrm{rot}}, \mathbf{p}_{\mathrm{el}}) \overset{f_5}{\mapsto} (\mathcal{R}_{\theta_{\mathrm{rot}}}, \mathbf{p}_{\mathrm{el}} ) \overset{f_6}{\mapsto} \mathbf{p'}$

If we graph the arc length of the boundary where each collision happens against the outgoing angle $\varphi$, we obtain the collision space.
We find a strange attractor that appears and changes as we vary the non-elastic parameter $\Lambda$. The structure of the attractor has been rigorously proved for $0 < \lambda < \frac{1}{3}$.

## Lyapunov exponents

To calculate the Lyapunov exponents of the billiard dynamics, we need to know how the tangent vectors $\delta\Gamma$ evolve.

When the particle is moving inside the region, the dynamics are linear and the evolution of the $\delta\Gamma$ vectors is simple.
The difficulty arises when the particle collides with the boundary.

Following the procedure of Dellago et al., the transformation of the tangent vectors at a collision is given by


\begin{equation}
\delta\mathbf{\Gamma'} =   D\mathsf{M}(\mathbf{\Gamma}) \cdot \delta\mathbf{\Gamma} + \left[   D\mathsf{M}(\mathbf{\Gamma})\cdot 
\mathbf{F(\mathbf{\Gamma})}  - \mathbf{F}(\mathsf{M}(\mathbf{\Gamma}))  \right] \delta\tau_c
\end{equation}

Here, $\mathbf{F}$ are the forces appearing in the equations of motion within the region, $\mathsf{M}$ is the map that corresponds to the collisions and $\delta\tau_c$ is the *time delay*, the time interval between the collision of the reference orbit and that of the satellite orbit.

Since the  collision only changes the momentum of the particle and not its position, the important part of the derivative $D\mathsf{M}$ of the map is given by the derivative of the momentum transformation map, $\mathcal{F}$.

We can express $D\mathsf{M}$ as

\begin{equation}
D\mathsf{M}= 
\begin{pmatrix}
\mathbf{1} \\
D\mathcal{F}(\Gamma)
\end{pmatrix}
\end{equation}

Numerically the Lyapunov exponents are aproximated by

\begin{equation}
\lambda = \dfrac{1}{t}\sum_{i = 1}^{n} \log \Vert \delta\Gamma(\mathfrak{t}_i) \Vert
\end{equation}

## Numerical procedure

We initialise 4 orthonormal 4-dimensional vectors. After every collision, we store their 
magnitudes and use the Gram--Schmidt method to re-orthonormalize them.


We store the magnitudes of the vectors after each collision in a list and after 3000000 collisions
we approximate the Lyapunov exponents.

```python
def transformacion_colision(self, mesa, frontera, time, dG, normas, p_antes, p_despues, theta, alfa):
    """
    Trasforma los vectores tangentes de acuerdo a la regla de colision de David y agrega a la lista que se le pase como
    normas la norma después del proceso Gram-Shmidt. t = tiempo de vuelo libre
    """

    t = mesa.tangentes[frontera]
    n = mesa.normales[frontera]

    p = p_antes

    y1 = p - 2.0*dot(p,n)*n
    y2 = array([y1[0], y1[1], y1[0], y1[1]])
    y3 = array([arctan(dot(y2[:2],t)/dot(y2[:2],n)), y2[0], y2[1]])
    y4 = array([(1- alfa)*y3[0], y3[1], y3[2] ])
    y5 = array([ cos(y4[0]), -sin(y4[0]), sin(y4[0]), cos(y4[0]), y4[1], y4[2]  ])
    y6 = dot(y5[:4].reshape(2,2) , y5[4:])
    
    m = dot(self.Df6(y5), self.Df5(y4))
    m = dot(m, self.Df4(alfa, y3))
    m = dot(m, self.Df3(y2))
    m = dot(m, self.Df2(y1))
    m = dot(m, self.Df1(n,p))

    delta_p = p_despues - p_antes
    
    dG_nuevo = []

    for dg in dG:

        dq1, dq2, dp1, dp2 = dg
        dq = array([dq1, dq2])
        dp = array([dp1, dp2])

        dot_nq = dot(n, dq)
        dq1_nuevo , dq2_nuevo = dq +  delta_p*dot_nq/dot(p_antes, n) # 2.0*dot_nq*n

        dp_nuevo = dot(m, dp)
        dp1_nuevo, dp2_nuevo = dp_nuevo

        dg_nuevo = array([dq1_nuevo, dq2_nuevo, dp1_nuevo, dp2_nuevo])
        dG_nuevo.append(dg_nuevo)

    dG = dG_nuevo

    return array(dG)
```
